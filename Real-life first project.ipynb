## Q2: what city has the highest number of sales?

```python
all
```

```python
# add city column first using apply method
def get_city(address):
    return address.split(',')[1]

# get state column


def get_state(address):
    return address.split(',')[2].split(" ")[1]
        

#all["City"] = all["PurchaseAddress"].apply(lambda x:x.split(',')[1]) :this works also
all["City"] = all["PurchaseAddress"].apply(lambda x:get_city(x) + "("+get_state(x)+")")
#all = all.drop(columns = "state")

all.head()
```

```python
# to determine state with hghest sales
# group by  sales
city = all.groupby("City").sum()
city

```

```python

#get x- value as cities
x_value = [city for city,df in all.groupby("City")]
plt.bar(x_value,city["Sales"])
plt.title("Sales for different cities",fontdict = {"fontsize":15})
#plt.xticks(range(1,13,1))
plt.xticks(x_value,rotation = "vertical")

plt.yticks()
plt.xlabel("City Name",fontdict = {"fontsize":12})

plt.ylabel("Sales[in Naira]",fontdict = {"fontsize":12})
plt.show()
```

## Q5:What products are most often sold together


```python
all.head()

```

```python
# create a new data frame 
# lets check for duplicated values
dp = all[all['OrderID'].duplicated(keep = False)]
dp.head(20)

# group same product with the same order ID
dp["Grouped"]=dp.groupby("OrderID")["Product"].transform(lambda x:",".join(x))
dp.head()


```

```python
# remove the same duplicate of order
dp = dp[["OrderID", "Grouped"]].drop_duplicates()
dp.head(100)
```

```python
# get final count
# import new libraries for this
from itertools import combinations
from collections import Counter

count = Counter()
for row in dp["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list,2)))
for key,value in count.most_common(10):
    print(key,value)
    












```

## Q6:What product is sold the most and why do you think so?

```python
all.head()
```

```python
# group by quantity
# group product
product_g = all.groupby("Product")
by_quantity = product_g.sum()["QuantityOrdered"]
products = [product for product ,df in product_g]

```

```python
#x_value = [product for product,df in by_quantity["Product"]]
plt.bar(products,by_quantity)
plt.title("Sales for different products",fontdict = {"fontsize":15})
#plt.xticks(range(1,13,1))

plt.xlabel("Product",fontdict = {"fontsize":12})
plt.xticks(products,rotation="vertical")
plt.ylabel("Quantity Ordered",fontdict = {"fontsize":12})
plt.grid()
plt.show()

# overlay 
```

```python
# plot the prices of the product alonside the initial graph
prices = all.groupby("Product").mean()["PriceEach"]

# overlay the second y-axis
fig,ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.bar(products,by_quantity,color = "b")
ax2.plot(products,prices,'g.-')
ax1.set_xlabel("Product")

ax1.set_xlabel("Product")

ax1.set_ylabel("Quantity Ordered",color = "b")
ax1.set_xticklabels(products,rotation='vertical')

ax2.set_ylabel("Price[in USD]",color ="g")
plt.show()
```
